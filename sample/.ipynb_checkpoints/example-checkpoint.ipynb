```python
from sklearn import datasets
from polyssifier import Poly
import numpy as np
import pandas as pd
from pandas.tools.plotting import scatter_matrix
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species");
X = np.array(iris)[:,:4].astype('float')
y = np.array(iris)[:,-1]
#y = pd.Factor(iris['species']).labels
#X = iris.drop('species').get_values()
#y
```

```python
# single class example
mask= np.logical_or(y=='versicolor', y=='virginica')
x2 = X[mask,:]
y2 = ((y[mask] =='versicolor')*1).ravel()
pol = Poly(x2,y2, n_folds=5)
pol.run()
pol.plot()
```

```python
pol = Poly(X,y, n_folds=5)
pol.run()
pol.plot()
```
